### Rename month names from Portuguese to English

```python
month_translation = {
    'Janeiro': 'January',
    'Fevereiro': 'February',
    'Março': 'March',
    'Abril': 'April',
    'Maio': 'May',
    'Junho': 'June',
    'Julho': 'July',
    'Agosto': 'August',
    'Setembro': 'September',
    'Outubro': 'October',
    'Novembro': 'November',
    'Dezembro': 'December'
}
df['month'] = df['month'].map(month_translation)
```

```python
df
```

### Which month maximum number of forest fires were reported and visualize them

```python
max_fires_month = df.groupby('month')['number'].sum().idxmax()
print("Month with maximum number of forest fires:", max_fires_month)
# To visualize
plt.figure(figsize=(10, 6))
sns.barplot(x='month', y='number', data=df, estimator=sum, ci=None)
plt.title('Total Forest Fires per Month')
plt.xlabel('Month')
plt.ylabel('Total Number of Fires')
plt.show()
```

### Which Year maximum number of forest fires were reported and visualize them

```python
max_fires_year = df.groupby('year')['number'].sum().idxmax()
print("Year with maximum number of forest fires:", max_fires_year)
# To visualize
plt.figure(figsize=(10, 6))
sns.barplot(x='year', y='number', data=df, estimator=sum, ci=None)
plt.title('Total Forest Fires per Year')
plt.xlabel('Year')
plt.ylabel('Total Number of Fires')
plt.show()
```

### Which states maximum number of forest fires were reported and visualize

```python
max_fires_state = df.groupby('state')['number'].sum().idxmax()
print("State with maximum number of forest fires:", max_fires_state)
# To visualize
plt.figure(figsize=(12, 6))
sns.barplot(x='state', y='number', data=df, estimator=sum, ci=None)
plt.title('Total Forest Fires per State')
plt.xlabel('State')
plt.ylabel('Total Number of Fires')
plt.xticks(rotation=45)
plt.show()
```

### Find the total number of fires reported in the Amazonas

```python
total_fires_amazonas = df[df['state'] == 'Amazonas']['number'].sum()
print("Total number of fires reported in Amazonas:", total_fires_amazonas)
```

### Display number of fires reported in Amazonas (year-wise) and visualize 

```python
amazonas_fires_yearly = df[df['state'] == 'Amazonas'].groupby('year')['number'].sum()
print("Number of fires reported in Amazonas (year-wise):\n", amazonas_fires_yearly)
# To visualize
plt.figure(figsize=(10, 6))
plt.plot(amazonas_fires_yearly.index, amazonas_fires_yearly.values, marker='o', linestyle='-')
plt.title('Number of Fires Reported in Amazonas (Year-wise)')
plt.xlabel('Year')
plt.ylabel('Total Number of Fires')
plt.xticks(amazonas_fires_yearly.index, rotation=45)
plt.grid(True)
plt.show()
```

### Display number of fires reported in Amazonas (day-wise) and visualize 

```python
amazonas_fires_daily = df[df['state'] == 'Amazonas'].groupby('date')['number'].sum()
print("Number of fires reported in Amazonas (day-wise):\n", amazonas_fires_daily)
# To visualize
plt.figure(figsize=(15, 6))
plt.plot(amazonas_fires_daily.index, amazonas_fires_daily.values, marker='o')
plt.title('Number of Fires Reported in Amazonas (Day-wise)')
plt.xlabel('Date')
plt.ylabel('Total Number of Fires')
plt.xticks(rotation=45)
plt.show()
```

### Find the total number of fires that were reported in 2015 and visualize based on each month

```python
fires_2015 = df[df['year'] == 2015]
total_fires_2015 = fires_2015['number'].sum()
print("Total number of fires reported in 2015:", total_fires_2015)
# To visualize
plt.figure(figsize=(10, 6))
sns.barplot(x='month', y='number', data=fires_2015, estimator=sum, ci=None)
plt.title('Total Fires Reported in 2015 (Month-wise)')
plt.xlabel('Month')
plt.ylabel('Total Number of Fires')
plt.show()
```

### Average number of fires reported from highest to lowest(state-wise) and visualize

```python
avg_fires_state = df.groupby('state')['number'].mean().sort_values(ascending=False)
print("Average number of fires reported (state-wise):\n", avg_fires_state)
# Visualize
plt.figure(figsize=(12, 6))
sns.barplot(x=avg_fires_state.index, y=avg_fires_state.values)
plt.title('Average Fires Reported by State (Highest to Lowest)')
plt.xlabel('State')
plt.ylabel('Average Number of Fires')
plt.xticks(rotation=45)
plt.show()
```

### Find the states where fires were reported in Dec month.

```python
december_states = df[df['month'] == 'December']['state'].unique()
print("States where fires were reported in December:", december_states)
```
